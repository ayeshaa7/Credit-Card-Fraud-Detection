```python
legit_sample = legit.sample(n=492)
```

```python
#Concatinating two data frames

new_dataset = pd.concat([legit_sample, fraud], axis =0) #axis=0 -> rows, and axis=1 -> columns of the data 
```

```python
#Spliting data into features (V1, V2, V3) and targets (0, 1).

X = new_dataset.drop(columns= 'Class', axis=1) #drops the class column
Y = new_dataset['Class'] #represents the class/labels
```

```python
print(X)
```

```python
#Class column dropped ^
```

```python
#Split data into Training Data and Testing Data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state =2) #x are the features/values of the 30 columns #y is the target variable/label/class, test size is 20% testing data we want. 80% is data is training data. 
#stratify is whne you is used to ensure that the split maintains the same proportion of different classes in both the training and testing sets as they are in the original dataset. 
#This is especially important when you have an imbalanced dataset, where some classes are more frequent than others.
#example if you have a dataset with 1000 samples where 900 belong to class 0 and 100 belong to class 1. 
#If you split this data without stratification, you might end up with a training set heavily skewed towards one class, 
#which can bias your model. By using stratify=y, we ensure that both the training and testing sets have approximately 90% of class 0 and 10% of class 1, reflecting the original dataset's distribution.



scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)


X_test_scaled = scaler.transform(X_test)

```

```python
print(X.shape, X_train.shape, X_test.shape) #984 values in the original data set of X, then X_train has 80% and X_test has 20%, we've split the data
```

```python
#Model Training  - through logistic regression model for binary classification problems.

model = LogisticRegression(max_iter=1000)
```

```python
#training the lgosistic regression model with the training data
model.fit(X_train_scaled, Y_train) #used fit to train the model. 
#then we will make predictions from this ML model
#we have lesser data points, in case of deep learning, we use more data points, training takes long time.
#this is numeric data, usually we have image data
```

```python
#finding performance of the model

#Accuracy Score on training data using X_train, and compare it to the original model in the Y_train
X_train_scaled_prediction = model.predict(X_train_scaled) #training all the labels for the training data, stored in this
training_data_accuracy = accuracy_score(X_train_scaled_prediction, Y_train) #then compare the values predicted by the model, with the original lavels stored in the Y-train
#stored in the training_data_accuracy
```

```python
print('Accuracy on Training data: ', training_data_accuracy)
#we test training data to check if the training and test data accuracy are very different; if they are different it means that our model is underfitted or overfitted.
#overfitting - model is doing well on training data - becuause model is overtrained on the training data as it has read it before.
#underfitted - over generalize learning
```

```python
#good score if more than 75. so 95/100 predictions are predicted of the training data by the model. 
#test data prediction is important. student studyign for math, practicing with problems. those problems will be trianing data. the examiner can ask the exam not present in the book.
#so test is evaluate with data it hasn't seen. it has seen the X_train data, but not the X_test data.
```

```python
X_test_prediction = model.predict(X_test_scaled)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data: ', test_data_accuracy)
```
